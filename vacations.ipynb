```python
# Dependencies and Setup

import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json

# Import API key
from api_keys import g_key
```

```python
# Store csv created in part one into a DataFrame
vac_data_df = pd.read_csv("VacationspotsUSA.csv")
vac_data_df
```

```python
# Configure gmaps
gmaps.configure(api_key=g_key)
```

```python
dest_df = vac_data_df[["Places", "Lat", "Long", "crime rate for county violent", "US average (violent)","Crime rate for county property","US average property"]].copy()
dest_df ["#restaurants"]= ""
dest_df ["ave_rest_rating"] =""
dest_df ["#attractions"]=""
dest_df ["ave_attraction_rating"]=""
dest_df ["icon"]=""
dest_df
```

```python
# Find Place ID
params = {
    "key": g_key,
    "inputtype": "textquery",
    "fields": "icon"
}

# Iterate through 
for index, row in dest_df.iterrows():
    # get place from df
    places = row["Places"]
   
    params["input"] = f"{places}"
    
    # Use the search term for each place

    base_url = "https://maps.googleapis.com/maps/api/place/findplacefromtext/json"

    # make request and print url
    response = requests.get(base_url, params=params)
    
    # convert to json
    place_ids = response.json()
    
    # Grab the place id from the results and store the name
    dest_df.loc[index, "icon"] = place_ids["candidates"][0]["icon"]
  
dest_df
```

```python
# Set parameters to search for tourist attractions
params = {
    "radius": 50000,
    "type": "tourist_attraction",
    "key": g_key
}

# Iterate through 
for index, row in dest_df.iterrows():
    # get lat, lng from df
    lat = row["Lat"]
    lng = row["Long"]
    
    params["location"] = f"{lat},{lng}"
    
    # Use the search term: "tourist_attraction" and our lat/lng
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # make request and print url
    attraction = requests.get(base_url, params=params)
    
    # convert to json
    attractions = attraction.json()
    
    count=0
    
    for x in attractions["results"]:
        if x["name"]:
            count+=1
    #print(count)
    dest_df.loc[index, "#attractions"] = count
    
    
    ratings=[]
    
    for x in attractions["results"]:
        #pprint(x)
        try:
            ratings.append(x["rating"])
        except:
            print("N/A")
    #print(ratings)
    attr=np.mean(ratings)
    
    dest_df.loc[index, "ave_attraction_rating"] = np.around([attr], decimals=1)

dest_df
    
```

```python
# Set parameters to search for tourist attractions
params = {
    "radius": 50000,
    "type": "restaurant",
    "key": g_key
}

# Iterate through 
for index, row in dest_df.iterrows():
    # get lat, lng from df
    lat = row["Lat"]
    lng = row["Long"]
    
    params["location"] = f"{lat},{lng}"
    
    # Use the search term: "tourist_attraction" and our lat/lng
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    # make request and print url
    restaurant = requests.get(base_url, params=params)
    
    # convert to json
    restaurants = restaurant.json()
    
    count=0
    
    for x in restaurants["results"]:
        if x["name"]:
            count+=1
    #print(count)
    dest_df.loc[index, "#restaurants"] = count
    
    
    ratings=[]
    
    for x in restaurants["results"]:
        #pprint(x)
        try:
            ratings.append(x["rating"])
        except:
            print("N/A")
    #print(ratings)
    rate=np.mean(ratings)
    
    dest_df.loc[index,"ave_rest_rating"] = np.around([rate], decimals=1)

dest_df
    
```
